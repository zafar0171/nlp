```python
from sklearn.feature_extraction.text import CountVectorizer

def extract_features(corpus):
    vectorizer = CountVectorizer(max_features=1500, min_df=5, max_df=0.7)
    features = vectorizer.fit_transform(corpus).toarray()
    return features
```

```python
def split_using_dictionary(text, word_dict):
    word_list = []
    current_word = ''
    for char in text:
        current_word += char
        if current_word.lower() in word_dict:
            word_list.append(current_word)
            current_word = ''
    return ' '.join(word_list)

# Example usage (you need a more comprehensive dictionary for real-world cases)
#word_dict = {'on', 'facebook', 'twitter', 'youtube', 'pinterest', 'company', 'careers', 'advertise', 'privacy'}
#text = "ONfacebooktwitteryoutubePinterestCOMPANYCareersAdvertisePrivacy"
#split_text = split_using_dictionary(text, word_dict)
#print(split_text)
```

```python
import re

def split_camel_case(text):
    matches = re.finditer('.+?(?:(?<=[a-z])(?=[A-Z])|(?<=[A-Z])(?=[A-Z][a-z])|$)', text)
    return ' '.join([m.group(0) for m in matches])

#text = "OnFacebookTwitterYouTubePinterest"
split_text = split_camel_case(text)
print(split_text)
```
