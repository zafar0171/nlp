```python
def cloud(df,category):
    splitted = ''.join([i for i in df_new[df_new['sentiment'] == category]['review_full']])
    wc = WordCloud(background_color='white', 
                   max_words=50, 
                   stopwords = stopword_list)
    plt.figure(figsize=(10,10))
    plt.imshow(wc.generate(splitted))
    plt.title('{} Sentiment Words'.format(category), fontsize=20)
    plt.axis('off')
    plt.show()
#twi = =" ".join([tweet for tweet in df[df.sentiment==category]['clean_text']])
```
